```python
def remove_punc(copydata):
  pattern = r'[' + string.punctuation + ']'
  copydata['product_name_fk']=data['product_name_fk'].map(lambda m:re.sub(pattern," ",m))
  copydata['product_name_az']=data['product_name_az'].map(lambda m:re.sub(pattern," ",m))
  return copydata


def lower(copydata):
  copydata['product_name_fk']=copydata['product_name_fk'].map(lambda m:m.lower())
  copydata['product_name_az']=copydata['product_name_az'].map(lambda m:m.lower())
  return copydata


def tokenization(text):
    tokens = re.split(' ',text)
    return tokens

def token(copydata):
  copydata['product_name_fk']= copydata['product_name_fk'].apply(lambda x: tokenization(x))
  copydata['product_name_az']= copydata['product_name_az'].apply(lambda x: tokenization(x))
  return copydata


sw=nltk.corpus.stopwords.words('english')

def remove_SW(copydata):
   copydata['product_name_fk']=copydata['product_name_fk'].apply(lambda x: [item for item in x if item not in sw])
   copydata['product_name_az']=copydata['product_name_az'].apply(lambda x: [item for item in x if item not in sw])
   return copydata


def remove_digits(copydata):
  copydata['product_name_fk']=copydata['product_name_fk'].apply(lambda x: [item for item in x if not item.isdigit()])
  copydata['product_name_az']=copydata['product_name_az'].apply(lambda x: [item for item in x if not item.isdigit()])
  return copydata


lemmatizer = WordNetLemmatizer()

def lemmatize(copydata):
  copydata['product_name_fk']=copydata['product_name_fk'].apply(lambda x: [lemmatizer.lemmatize(item) for item in x])
  copydata['product_name_az']=copydata['product_name_az'].apply(lambda x: [lemmatizer.lemmatize(item) for item in x])
  return copydata


def remove_empty_tokens(copydata):
  copydata['product_name_fk']=copydata['product_name_fk'].apply(lambda x: [item for item in x if item !=''])
  copydata['product_name_az']=copydata['product_name_az'].apply(lambda x: [item for item in x if item !=''])
  return copydata


def remove_single_letters(copydata):
  copydata['product_name_fk']=copydata['product_name_fk'].apply(lambda x: [item for item in x if len(item) > 1])
  copydata['product_name_az']=copydata['product_name_az'].apply(lambda x: [item for item in x if len(item) > 1])
  return copydata


def detoken(copydata):
  copydata['product_name_fk']= copydata['product_name_fk'].apply(lambda x: TreebankWordDetokenizer().detokenize(x))
  copydata['product_name_az']= copydata['product_name_az'].apply(lambda x: TreebankWordDetokenizer().detokenize(x))
  return copydata

def replace_spaces(x,space,second):
  result = x.replace(space, second)
  return result
def remove_space(copydata):
  copydata['product_name_fk']= copydata['product_name_fk'].apply(lambda x: replace_spaces(x,'  ',' '))
  copydata['product_name_az']= copydata['product_name_az'].apply(lambda x: replace_spaces(x,'  ',' '))
  return copydata
def count_vcr():
  for i in range(len(copydata)):
    doc1=copydata['product_name_fk'][i]
    doc2=copydata['product_name_az'][i]
    docs=(doc1,doc2)
    matrix = CountVectorizer().fit_transform(docs)
    cosine_sim = cosine_similarity(matrix[0], matrix[1])
    similarity.append(cosine_sim)
  return similarity

def similarity_fn():
  for i in range(len(copydata)):
    doc1=copydata['product_name_fk'][i]
    doc2=copydata['product_name_az'][i]
    docs=(doc1,doc2)
    tfidf_matrix = tfidf_vectorizer.fit_transform(docs)
    cosine_sim = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])
    similarity.append(cosine_sim)
  return similarity
```
